## 阅读的层次
### 基础阅读
- 达到有基本阅读能力的水平

### 检视阅读
- 有系统的粗读/略读
  - 书名
  - 目录页
  - 索引
  - 出版者介绍
  - 注意最后三页
  
- 粗浅的阅读
  - 通读一遍整本书
  - 略过不懂的地方
  - 不要停留和查阅资料
  - 1 ~ 2 小时完成
  
- 回答作为读者的问题
  - 4个问题
  - 是否有必要精读

### 分析阅读 (P175 总结)
#### 阶段1 —— 整理大纲（P106）
- 书本分类
  - 实用性：怎么用
  - 理论性：是什么
- 几句话概括本书
- 列举重要章节、说明如何组成整体架构
- 找出作者在问的问题、或者作者想要解决的问题

#### 阶段2 —— 诠释内容（P148）
- 找出关键字、与作者达成共识
  - 诠释内容和讯息
  - 让你头痛的字
- 将书中最重要的句子圈起来，找出其中的主旨
- 从相关文句关联、设法架构出一本书的基本论述
- 找出作者的解答
  - 确定作者解决了哪些问题，还有哪些未解决
  - 未解决的问题中，哪些是作者认为自己无法解决的
  
#### 阶段3 —— 批评式阅读（P163）
- 完整的了解一本书，不要急于批评
- 不要争强好胜或盲目反对
- 将知识上不同意见看作是大体上可以解决的问题
  - 读者要为自己不同的意见找理论基础才有机会解决这个问题

### 主题阅读（P349）
- 第一个要知道：一个特定的问题，所涉及的绝不是一本书
- 第二个要知道：应该读哪些书

#### 检视阅读在主题阅读的作用
- 检视书单上所有的书
  - 让你对自己想要研究的主题有个清晰的概念
  - 简化你的书目到一个合理的程度
  - 这本书谈的内容对他研究的主题到底重不重要

#### 主题阅读的五个步骤
1. 找到相关的章节
  - 主题阅读中，你关心的主题才是基本的重点，而不是你阅读的书

2. 带引作者与你达成共识
  - 由你自己来建立共识

3. 厘清问题
  - 按照顺序列举一些问题，然后让作者来做出回答
  
4. 界定议题
  - 如果作者给出的只有正反两面的意见，这个问题算是简单问题
  - 通常会有超过两种以上的答案，这时需要找出不同意见彼此之间的关联，再根据作者的观点来作分类

5. 分析讨论
  - 依照特定顺序来提问，并能够辨认为什么是这个顺序
  - 必须

## 不同读物阅读技巧
### 想象文学
- 以达成某种深沉的经验为目标
- 教导我们衍生的意义
- 规则
  - 整本书的大意
  - 发现各个部分是如何架构起来的

### 故事、戏剧与诗
- 故事书
  - 快读
- 戏剧
  - 不完整
  - 无法直接与读者对话
  - 把自己融入演出中
- 诗
  - 一口气读完，不要停
  - 重读一遍，大声读出来
  - 诗中所谈论的冲突：爱与时间、生与死、短暂的美与永恒的胜利

### 历史
- 都有一个特殊而且有限定范围的主题
- 整书是按照什么方式来区分不同章节的？年代、时期或世代
1. 对你感兴趣的事件或时期，尽可能阅读一种以上的历史书
2. 阅读历史时，不要只关心在过去某个时间、地点真正发生了什么事，还要读懂为什么会有如此这般行动的原因，尤其是现在

### 传记与自传
- 定案本
  - 不是用来写活人的
- 授权本
  - 由继承人或者重要人物的朋友来负责

### 科学与数学
- 了解科学这门行业
- 把数学当成一种语言
  - 不要头痛书中的数学问题

### 哲学
- 哲学就是一种思考
- 要回答哲学问题，只有思考
- 论述风格
  - 哲学对话
  - 哲学论文或散文
  - 哲学系统化
  - 格言形式
  
- 时间线
  - 苏格拉底
    - 柏拉图
      - 亚里士多德
      - 康德
      - 阿奎那
      - 尼采
      - 笛卡尔/斯宾诺莎

### 社会科学
- 什么是社会科学
  - 通常包括了人类学、经济学、政治学、社会学
  - 专注于追求人类社会的系统知识

## 概述
- 在执行过程中，容易半途而废的一个很可能的原因在于节奏出了问题

## 计划的节奏
- 长期
  - 一个长期的目标可以是五年，也可以是十年，因目标而异。

- 中期
  - 一个中期的目标，也许是一年
  - 比如，计划用一年时间写一本书，假如一本书 20 万字，那每周大约需要完成 4000 字，再细化到每天就是 800 字左右。
  
- 短期
  - 可能是几个月
  - 比如这个专栏：它上面已经形成了每周三篇更新的节奏，这样的写作节奏对于我来说基本已经算是全力冲刺了，所以时间就不能拉得太长。

- **不同周期的计划，都会有一个共同的问题：计划总是过于乐观了，现实的执行很难完全符合计划。**

### 怎么应对意外
- 为所有计划中的事情创建了一个优先级队列，每次都只取一件最高优先级的事情来做。
- 现实总会有临时更高优先级的 "意外" 紧急事件插入，处理完临时的紧急事件，队列中还排满着很多本来计划当天要做的事情。
  - 如果计划被意外打断了，就跳过，不用太在意，明天继续把它加入到队列中继续执行即可。

### 他人的节奏
- 体会：每个人都会有自己不同的节奏，这需要自己去摸索、练习，并慢慢提升。
  - 节奏太快：会疲惫、倦怠，很容易放弃；
  - 节奏太慢：达不到练习与提升的效果，浪费时间
  
### 自己的节奏
- 找到并控制好自己的节奏，才能长期匀速地奔跑，才能更高效地利用好自己的时间和注意力。
- **把自己的时间安排成一段一段的，高度集中和高度分心交叉分布**
  - 高度集中注意力：思考处理比较难的，远离手机等环境干扰；然后休息 15 分钟左右，回复消息之类的琐碎事项
  - 外部干扰：很难消除，则不必勉强，调整环境情况再继续
  - 疲惫：小憩片刻、闭目养神，恢复好后再继续

- **找到自己的节奏，就是在每天略感挑战的状态下，形成不断加速前行，直到一个最终接近匀速的状态。匀速是我们能长期坚持的临界点，它能让我们跑得更久，跑得更远。**

# 01. HTTP 到底是什么
- 游览器输入地址，打开网页
- Android 中发送网络请求，返回对应数据

> HyperText Transfer Protocol: 超文本传输协议

---

## 02. HTTP 工作方式
- 请求 -> 响应

---

## 03. URL -> HTTP 报文
- 示例：http://demourl.com/user?gender=male
  - 协议: http://
  - 服务器地址: demourl.com
  - 路径: /user?gender=male

---

## 04. 请求格式：Request
```
- 请求行   method      path       HTTP version
            GET      /users/1      HTTP /1.1
- Headers  Host: api.github.com
           Content-Type: text/plain
           Content-Length: 123
- body     bodybodybodybodybodybodybodybody
           bodybodybodybodybodybodybodybody
```

---

## 05. 响应格式：Response
```
-             HTTP version    status code     status message
- 状态行         HTTP/1.1         200             OK
- Headers       content-Type: application/json; charset=utf-8
                cache-control: public, max-age=60
                ......
                content-encoding: gzip
- body          [{"login":"peter","id":1, .......}]
```

---

## 06. 关键内容
### 6.1. Request method
- GET: 获取资源；没有 body
  - 幂等
- POST: 增加或者修改信息; 有 body
  - 非幂等
- PUT: 只修改资源; 有 body
  - 幂等
- DELETE: 删除资源; 没有 body
  - 幂等
- HEAD: 基本等同于 GET
  - 不会返回 body
  - 比如下载的时候，可以知道文件的大小，而不会直接下载文件

### 6.2. Response status code
- 1xx: 临时性消息
  - 100: Continue(迄今为止的所有内容都是可行的，客户端应该继续请求)
  - 101: Switching Protocol
- 2xx: 成功
  - 200: 成功
- 3xx: 重定向
  - 301: Moved Permanently(被请求的资源已永久移动到新位置)
  - 302: Found(请求的资源现在临时从不同的 URI 响应请求)
  - 304: Not Modified(内容没有改变)
- 4xx: 客户端错误
  - 400: Bad Request
  - 404: Not Found
- 5xx: 服务器错误
  - 500: Internal Server Error
  - 503: Service Unavailable

### 6.3. Headers
- 传输元数据(metadata)
  - Host
    - 服务器主机地址
    - 仅用于找到目标主机后确认主机域名和端口
  - Content-Length
    - 内容的长度(字节)
    - 内容可能是二进制的内容，无法通过分隔符来确定内容结束
  - Content-Type
    - text/html: html 网页
    - application/x-www-form-urlencoded: 普通表单，encoded URL 格式
    - multipart/form-data: 多部分形式，一般用于传输包含二进制内容的多项内容
    - application/json: json 形式，用于 web api 的响应或 post/put 请求
    - image/jpeg 或 application/zip ...: 单文件，用于 web api 响应或 post/put 请求
  - Chunked Transfer Encoding
    - Transfer Encoding: chunked
    - 服务器已经拿到一部分内容了，还有一部分比较耗时，需要再等等，所以服务端会分批下发
  - Location
    - 重定向的目标 URL
  - User-Agent
    - 用户代理
  - Range/Accept-Range
    - 指定 Body 的内容范围
  - Cookie/Set-Cookie
    - 发送 Cookie / 设置 Cookie
  - Authorization
    - 授权信息

- 部分其他 Header
  - Accept
    - 客户端能接受的数据类型
  - Accept-Charset
    - 客户端接手的字符集
  - Accept-Encoding
    - 客户端接手的压缩类型
  - Content-Encoding
    - 压缩类型，如 zip

#### Cache
- Cache 和 Buffer 的区别
  - buffer 针对工作流的: 视频缓存
- Cache-Control
  - no-cache
    - 每次使用的时候询问下
  - no-store
    - 不需要缓存
  - max-age
    - 失效时间之前不需要询问，直接用
  - private/public
    - 定制化的个性信息
- Last-Modified
  - If-Modified-Since
- Etag
  - If-None-Match

####  REST
- 是一种架构风格
  - Server-Client 架构
  - 无状态
  - 可缓存
  - 分层系统（集群）
  - 可在响应中提供代码
  - 统一的接口

- RESTful HTTP
  - 正确使用 HTTP

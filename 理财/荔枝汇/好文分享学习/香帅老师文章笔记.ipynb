## 银行躺着赚钱的秘密（20190102 - 得到：北大金融学课 016）
### 1. 存贷差赚钱
#### 利润
- 2016 年中国银行实现的净利润大概是 1.6 万亿元，其中存贷差差不多占到 3/4 左右，其他 1/4 利润主要来源于：
  - 存款自投资：比如国债（比较安全的资产）
  - 通用业务费用，比如银行理财产品，其实都是和其他金融机构合作的，银行主要是收取通道费
- 三块利润：存贷差、自投资业务、通道费，依赖于银行的规模和吸收存款的能力

#### 银行作用
- 银行通过资金规模、期限和风险转换，实现资金更优配置，赚取利差
- 规模转换
  - 银行通过收集个人存款，集中后形成规模，再借贷给资金量需求大的人
- 期限转换
  - 银行将不同时间点收集上来的资金，匹配给资金使用期限不同的人，比如一年期、两年期、五年期等，打一个时间差
- 风险转换
  - 银行需要承担贷款收不回来的风险

### 2. 躺着赚钱的原因：银行的垄断性和高速的经济增长
- 准入门槛高，牌照控制很严格
- 国家处于高速增长阶段（2003 ~ 2013年）

### 3. 银行业贫富分化的新业态：利率市场化和经济增速下滑
- 银行准入门槛被降低了。（2000年时候只有 40 家银行，现在有 4000 多家银行）
- 2013 年后，国家开始推行利率市场化，存款利率的管制放开了

## 投资银行：不做银行业务的银行（得到：北大金融学课 026）¶
### 1. 券商其实就是投行
- 帮普通老百姓开户，赚取手续费，叫 "零售经济"，在全球投行业里面，占比大概连 10% 都不到
- 中国的券商，主要依靠零售经济，占比达到 50%。

> 投行包含的业务：
  - 公司发行股票、债券，叫做：证券承销
  - 企业要搞兼并收购，需要人牵线搭桥，进行评估，提供融资安排建议，叫做：兼并收购
  - 其他包括：资产证券化、杠杆并购、私募股权投资、量化投资
  - 国际上：发行自行业务是做得最多的，大概能够占到 70% 以上的比重，而发行、承销、兼并、收购，大概占比 20% ~ 30%，零售经济占比是非常小的

### 2. 投行和银行的关系
- 早期世界上并没有投资银行的概念，基本都是即吸收储蓄、发放贷款，又做证券承销业务的金融寡头
  - 问题：储户希望得到固定的回报，但是银行把收到的钱去做证券业务，导致风险不对等，也没有保护，容易导致金融危机
- 1933 年版本《格拉斯.斯蒂格尔》法案
  - 最重要的一条：商业银行业务和证券业务要严格分开，实施分业经营

### 3. “已婚男士”与“快乐的单身汉”
- 商业银行，因为承接的是老百姓的储蓄，需要保证储蓄的安全，所以监管比较严格；而投资银行，投资者需要自己承担风险的，所以监管一直比较松；
  - 比喻：商业银行像 "已婚男士"，生活有规律，三餐有人管；投行更像是 "快乐的单身汉"，约束少，容易创新，但也容易冒险闯祸。
  - 创新：比如货币基金、资产证券化、杠杆并购、私募股权投资、风投，创投

## 谷歌上市记（20190119 - 得到：北大金融学课 029）
### 1. 不同寻常的 IPO
- google 召集了 10+ 公司来统计沟通，填写问卷调查，从而自己来筛选投行，而不是像传统的那样自己去拜访投行公司，类似于：皇帝选妃
  - google 为这次的承销支付的费用很高，达到 1亿美元
  - google 的名气大，投行公司不想错过
  
- google 的定价比价任性，自己的期望价格达到 108 ~ 135 美元，在当时市场上是对标公司的 7 倍之多，最后 google 做了一定的让步，但也达到 85 美元。（一般是由投行去询价，然后给 IPO 定价）
- 定价方式上，google 也别具一格，使用了一种叫："荷兰式拍卖"的方式
  > 荷兰式拍卖
  - 1. 上市公司自己发表公告，宣布发信股票的最大股数和股票潜在的价格区间
  - 2. 投资者看到这个公告以后开始同时提交订单，然后写上愿意以多少价格买多少股票
  - 3. 根据由高到底的拍卖方式，出价最高的投资者优先可以购得股票，最后结算价格是将所有的股票全部卖出的价格（如发行 100 股，a 竞价 90 美元 80 股，b 竞价 88 美元 80 股，则优先以 90 美元成交 80 股卖给 a，剩下的 20 股以 88 美元成交给 b，最终的结算价格就是 88 美元）


### 2. 资本为王时代的结束
- 钱越来越不值钱的时代，原来的世界是资金贵、资金稀缺
- 人力资源最关键，企业家精神最关键，所以企业和企业家的话语权越来越强，而且越是科技含量高、人力资源集中的行业，企业的主动权就越大。

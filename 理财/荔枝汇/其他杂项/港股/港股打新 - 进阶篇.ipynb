## 1. 招股价
- 招股价就是上市公司初步制定的股票发行价，招股价往往是一个价格区间。
- 最终确定为多少，要结合订单情况，市场情况，发行人的意愿等各种因素确定。
- 公司上市发行价一般位于招股价的价格区间。

## 2. 卖出时间
- 港股没有涨跌停板限制，一般都可以迅速反应利好或利空消息
  - 如果是涨，千万别幻想未来涨更多，咱们没法判断未来是涨还是跌，赶紧卖出保住利润；
  - 如果是跌，那更不能恋战，很多人一看到浮亏卖出心里亏得慌，再亏得慌也得卖，要不然可能未来会亏更多

- 时间：**建议一定要在新股上市第一天的早上卖出，更准确点就是在首日的 11 点或 10 点前卖出**
- 打新就是打新，打新就是为了卖，不是为了当股东，切勿恋战！切勿恋战！

## 3. 暗盘
- 暗盘交易，就是通过某些大型券商自行设立的内部系统进行报价撮合，目前暗盘做得最大的是辉立证券，辉立暗盘的成交量占全部暗盘成交量的90%左右，大部分券商都是接入辉立的暗盘系统。（利弗莫尔是自己做的暗盘，只针对自己的客户，使用人数少，成交量也不高）
- 港股新股暗盘交易时间一般为新股正式上市日前一个交易日下午 4 时 15 分至 6 时 30 分。

![image](暗盘举例.png)

- 暗盘交易的好处在于，可以进行提前交易买卖
- 手续费：正常交易佣金 1 港币/笔，暗盘交易佣金 11.9 港币/笔
- **慎用：暗盘的另一个好处就是对于未中签的新股又给了你一个入场的机会**
- 支持暗盘的券商：***尊嘉、富途、华盛通、玖富**

![image](招股书.png)

## 6. 行情延迟
- 一般都会延迟 15 分钟，延迟 15 分钟是什么意思呢？当 14:05 的时候，我们在软件上依然只能看到截至到 13:50 的 K 线图，也就是少了 15 分钟的 K 线走势
- 思路：
  - 开户的时候系统就会自动送你几个月的"高级行情"福利，使用这个高级行情就能看到实时的 K 线走势了
  - 如果你不用技术分析，其实少这15分钟的K线图也没多大关系，港股打新基本上市首日就卖了

## 7. 绿鞋机制
- 绿鞋机制是干嘛的？稳定股价的
  - 绿鞋机制其实就是超额配售权的的俗称，就是多配售了一点股票，超额配售权最早是在1963年在美国绿鞋制造公司使用的，所以大家就称为绿鞋机制。

- 超额配售就是指承销商可以比原计划多发行 15% 的股份，本来计划发行 100 股，那承销商有权利再多发行 15 股（15%）。
  - 一般来说，大部分承销商会直接选择多发行 15% 的股份，如果上市首日，新股破发，那么承销商就可以用多发行那 15% 募集的资金从二级市场买入超额发行的那些股份，然后以发行价卖给基石投资者，这样一来承销商不就赚差价了；
  - 反之，如果上市首日新股暴涨，那多发行 15% 的股份，供给多了，也能给股价降降温

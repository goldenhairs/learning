## 为什么要港股打新
- 如果你想找一个海外的投资市场，而且大概率能赚钱的，那港股打新还是不错的

## 港股打新到底能拿到多高的年化收益？
- **不知道**
- 能知道的是历史上港股上市首日大概 70% 的新股是上涨的，30% 是下跌的，上涨的概率大于下跌的，就像你买了一堆奖券，70% 的能刮出钱，30% 亏钱。
- 还有一个很重要的事就是中签率，同样申购一只新股，有人中了，有人没中，那收益肯定是千差万别。
- 如果非要说个收益的话那大概在 20%--30% 左右，运气好的能更高，运气差的可能更低

## 什么样的人适合港股打新
- 有闲钱
  - 港股对交易时间的要求还是蛮高的，最起码你得先打新，然后还得留意在新股上市的时候卖出，因为你得经常打新、卖出，所以一定得有空余时间的会员再考虑。
- 有时间
  - 如果你平时银行账户里就躺着几万块没有用，那可以通过港股打新把这笔资金给利用起来

## 港卡和出入金
- 入金用港卡、国内卡都行（国内卡失败概率较高），但出金只能出金到境外银行卡里。
- 港卡搞定下一步就是入金了，入金流程稍微麻烦一点，你需要先换汇，把人民币换成港币，然后再入金。

### 入金
- [入金知识](https://www.zinvestglobal.com/qa/second/002)

### 入金多少比较合适
- 2-3 万港币，折合人民币在 1.7 万 - 2.6 万左右
- 打新资金的利用率还是蛮高的，一笔资金我们打新，中签然后卖出，再拿卖出得到的钱再打新，然后再卖出....这笔钱是在账户、在不同的新股中不断滚存的。另外打一手新股的资金大概需要 2 千-- 5 千左右，那么 2-3 万港币大概能同时申购 5-7 只新股，基本上也够了。

## 打新流程
- 顺便说说 A 股打新和港股打新的区别，A 股只有 1 天的时间给我们打新，如果今天没打，那就没有机会了，港股往往会有五六天的打新时间，你在这个打新周期内的任意一天打新都行
- 另外就是A股得持有股票市值才能打新，持有市值越高中签率越高，港股不需要持有市值，只要有账户、有资金就能打新。

## 中签率/我该申购几手
- **每支新股打新一手**
- 本身港股打新的中签率就高很多，之前打新一手的中签率能达到百分之八九十。这两年稍微降了降，17 年的中签率在 73%，18 年在 61% 左右

## 多账户打新
- **如果在港股重复打新，将会被港交所取消本次中签资格**
- 不过大陆和香港之间有个制度 BUG，大陆身份证有 18 位，香港只有 8 位，这就导致在个人身份识别上会出现问题，而且不同的证券公司提交给港交所的信息还不一样，比如华盛通是以公司名义整体提交，并不涉及个人信息；玖富是提供身份证前 8 位，尊嘉是提供后 8 位，所以同一人开通了玖富和尊嘉，即使重复打新也有效，因为在港交所眼里这是两个人。
- **千万记得不要用尊嘉和富途一起申购，因为他们都提交身份证后8位**
- 如果想多账户打新最好还是用多人身份开户

## 1. 香港股市简史
- 1870 年左右，香港的股票交易还没有专门的交易所，而是股票经纪人代投资人交易股票，佣金是 0.25% 的水平（相比之下，我们今天万分之一也就是 0.01% 真是太幸福了），那时候的股票市场跟赌场区别不大，投机成风，而且充斥着出老千的内幕交易行为，比如我公司明天要发新产品，那我和我家人先买进股票，等产品一出股价大涨我们就能赚钱了，比现在不知道混乱多少倍，基本就是蛮荒时代。
- 1891 年，香港股票经纪会成立，由会员每天在会所内提出报价，合适就成交，交易向政府登记。并且会把股票成交价格刊登在第二天的报纸上，供投资者参考，这里可以看到股票交易变得有序和透明了一些。
- 1947 年，香港股票经纪会和后来成立的香港股票经纪协会合并，才正式有了香港证券交易所，此后香港又诞生了远东交易所、金银交易所和九龙交易所，合称"四会"。1986 年，四所合并，香港联合交易所成立，并在 1989 年成立了证券及期货事务监察委员会，即现在的证监会。
- 1992 年，当时的主席李业广在上市公司匮乏的情况下，建议内地国企在香港上市，并在 93 年开始越来越多的内地企业成功在香港上市了。
- 1994 年，恒生指数服务公司推出了中国企业指数，来跟踪这些在香港上市公司的股价走势，简称国企指数或 H 股指数，也就是咱们今天经常纳入投资组合的 H 股 ETF 所跟踪的指数
- 2000 年 3 月，为提高香港竞争力，香港联合交易所又和香港期货交易所有限公司还有香港中央结算有限公司合并，由单一控股公司香港交易所拥有。这才基本成为了我们今天看到的香港交易所。

## 2. 港股入门 ABC
- 1. 港股实行 T+0 交易制度，当天买进当天卖出，而 A 股当天买进是不可以当天卖的。港股还实行 T+2 交收制度，就是说当天买入股票在第 2 个交易日以前，不能享受相关权益。比如股票明天要分红（股权登记日），那今天买入这张股票就不能拿到分红。
- 2. 港股没有涨跌限制。有时候会看到一只港股一天下跌 70%，80%，但那是 10 亿以下港币的小盘股比较容易发生的事情，蓝筹股每天的涨跌幅跟 A 股是差不多的。所以新手如果要买个股，建议先只接触蓝筹股，否则心脏可能受不了。
- 3. 港股 1 手多少股可以由上市公司自己定，A 股 1 手是固定的 100 股。比如腾讯 1 手是 100 股，而中国平安是 500 股。
- 4. 港股传统交易软件是涨是绿色，跌是红色，跟美国一样。但现在很多券商已经可以自定义设置颜色了。股票最小变动单位 A 股是固定的 0.01 人民币，港股是 0.001 港元到 0.2 港元不等
- 5. 看港股行情不少券商是收费的。港交所是上市公司，既然上市就要赚钱为股东谋求利益，所以港交所授权券商的实时行情是收费的。但不少互联网券商已经免费提供行情了，这部分授权费由它们自行承担了。
- 6. 香港休市的节日跟 A 股有相似也有很多不同。相同的有元旦（的第二天），农历新年初三初四，清明节，劳动节，端午节，国庆节（的第二天），中秋节（的第二天）。此外还有耶稣受难日，复活节，佛诞节，圣诞节和圣诞节的第二天。此外香港经常遇到台风，如果遇到 8 号以上台风港交所也会暂停交易让大家不用上班了，所以有些荔枝会发现不是什么节假日香港也莫名其妙休市了。
- 7. 交易费用。港股的交易费用很多很杂，除了券商的佣金各家不同，其他还有 5 项统一的收费：特区政府收的印花税，证监会收的交易征费，香港交易所收的交易费，香港结算所收的交收费和交易系统使用费。其中大头是佣金和印花税，佣金 0.03% 到 0.1% 不等（最低 3 港元到 100 港元不等，近期才出现免佣金），印花税 0.1% 双向征收，衍生品免印花税。
- 8. 港股连续竞价时间为交易日的上午 9：30 到 12：00，下午 13：00 到 16：00，可以看到比 A 股会稍长一些。此外还有集合竞价时间早上 9：00 到 9：20，下午 16：00 到 16：10，不建议在这段时间参与交易，股价容易大涨大跌。

## 4. 港股打新的知识点
- 香港的 IPO 实行注册制，但比美国的注册制要严格
- 具体上市流程是，公司向港交所上市科和香港证监会提交 IPO 资料双重存档。再来，是港交所上市委员会对 IPO 进行把关。同时，香港证监会发现 IPO 资料有作假或者误导陈述，可以叫停上市。之后，拟上市公司会在上面提到的港交所"[披露易](www.hkexnews.hk/index_c.htm)的"申请版本，及聆讯后资料集"发布招股说明书。并在通过上市聆讯后一两周内我们就可以申购新股。
- 港股打新中签率高，常常打 1 手也能中，港股打新不需要跟 A 股一样持有市值。但是和 A 股不同，不是打中新股就一定能赚钱的，所以我们一定要注意打新的一些事项。
  - 周期股不能随便打。比如券商股，行情好的时候发行定价高，上涨空间有限，行情不好的时候市场不认，容易亏钱。
  - 按照 1 倍 PB 发行的内地银行股、地产股，比二级市场的定价要高，容易破发，不建议打。
  - 日本、新加坡、欧洲等海外上市公司。历史上看海外上市公司表现都不怎么样，不建议打。
  - 纺织服装类股以往表现不佳，不建议打
  - 5 亿港币以下市值的小上市公司不建议打，风险较大
  - 相比同行业已上市公司定价过高的股不建议打

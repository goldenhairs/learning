## 详解消费指数
- 有些行业本身就是含着金钥匙出生的，市场上存在比较稳定的需求，那赚钱比较容易，比如消费指数，医药指数

### 行业分布
- 1999 年，摩根士利丹和标普联合推出了全球行业分类标准（GICS），将行业分为10个大行业，24个二级行业，和67个子行业，并建立了行业指数
- 我国的行业指数也是参考这个标准的，最主要的 10 个大行业就包括：
  - 医药、必需消费、可选消费、金融、材料、能源、工业、信息、电信、公共事业

![各行业收益](各行业收益.png)

- 300 行业指数是 2004 年底发布的，都是从 1000 点开始。
- 这 10 大行业中共有 4 个行业跑赢了沪深 300，分别是：必需消费、医药、金融、可选消费
  - 必需消费（主要消费）包括：食品、药品、家具用品，是维持日常生活的刚需品
  - 可选消费包括：汽车、零售、媒体、房地产等等
    - 必需消费（也叫主要消费）行业是指日常生活中最基本、最必要的消费，受经济周期波动的影响比较小，也就是属于弱周期行业
    - 可选消费行业是除了主要消费行业之外的其它各类消费，可以认为是各种升级的消费需求，受经济周期波动影响较大

### 选择哪些消费指数
#### 必需消费
- 上证消费
  - 由上海证券市场主要消费行业股票组成，上证消费的基准日期是 2003年12月31日，起点 1000 点，到19年1月份是 7817 点，折合年化 14.68%
- 中证消费
  - 由中证 800 指数样本股中的主要消费行业股票组成（中证 800 就是沪深 300 + 中证 500），中证消费的基准日期是 2004年12月31日，起点 1000 点，到19年1月份是10443点，折合年化 18.21%

> 中证消费的覆盖面更广，规模也更大
  - 上证消费 ETF 的规模只有 1.6 亿，场内一天的交易额只有几十万，交易并不活跃
  - 中证消费 ETF 的规模有 15 亿，场内一天的交易额在 1 千万左右，交易比较活跃
  - 必需消费建议考虑中证消费，场内代码：**159928**，场外代码：**000248**

#### 可选消费
- 全指可选指数
  - 从中证全指样本股可选消费行业内选择流动性和市场代表性较好的股票构成指数样本股（中证全指由全部A股股票中剔除ST、\*ST股票，以及上市时间不足3个月等股票后的剩余股票构成样本股）
  - 全指可选的基准日期是 2004年12月31日，起点 1000 点，到19年1月份是 3842 点，折合年化 10.09%
  - 收益不如必需消费：一个是因为本身必需消费的表现就要比可选消费好，另一个原因是目前可选消费的估值要比必需消费低很多，这也拉低了全指可选的历史收益
  - 全指可选的代码，场内代码：**159936**，场外代码：**001133**
  - 全指可选的规模并不大，ETF 规模在 2.2 亿，场内交易额很低，一天只有几万，交易并不活跃
  
#### 是否值得入手（2019年1月）
- 中证消费的 PE 分位点在 31.04%，处于合理偏低的位置，现在能不能投，直白说能投，不过这个估值其实不算非常低估，尤其是与其他指数相比
- 全指可选的分位点就比较低了，目前只有 3.24%，可以说全指可选绝对处于低估状态了，不过我不建议定投全指可选，原因很简单，从历史走势来看**医药**的收益是要好于**可选消费**的

## 股市税收
1. 交易过程中，征收的印花税，目前是对卖出方按交易额的 0.1% 征收
2. 股票红利税，这个大家相对比较熟悉，和基金分红收取的税费类似
   - 持股超过 1 年的，税负为 5%；持股 1 个月至 1 年的，税负为 10%；持股 1 个月以内的，税负为 20%
3. 资本利得税，二级市场交易的盈利部分暂不征收（征收了也没有鸡儿用，股民有几个盈利的？如果对亏损投资者进行补贴，估计证券市场市场会破产。在外国股市，这部分税费是免不了的）

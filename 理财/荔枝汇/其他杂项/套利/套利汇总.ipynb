## 1. 套利的概念
- 假如你去集市上买东西，在集东头一个苹果卖 10 块钱，而在集西头你发现同样的苹果却只卖 8 块钱。那么肯定就会想到一种赚钱的方法：花 8 块钱从西头买苹果，然后在运到东头以 10 块钱的价格卖出，那么每个苹果你就能赚到 2 块钱，盈利 25%！
> 所谓套利就是找到便宜的苹果，买便宜的苹果，然后以更高的价格卖出

## 2. 套利基金
- LOF 基金又叫上市型开放式基金，它最大的特点就是既可以在场内交易也可以在场外交易
- 既然在两个市场交易，那么就很容易出现两个不同的价格（这个说法其实是不严谨的，准确来说，场外的叫净值，场内的才叫价格），只要价格差距够大，那就有套利的空间
- 华宝油气就是做套利的一个良好标的，华宝油气既是 **QDII** 也是 **LOF**，也就是我们既能在华泰中买卖，也能从支付宝等场外平台申购赎回

## 4. 套利注意点
### 4.1. 手续费
- 一般来说，LOF 基金的申购费是 1.2%-1.5%，指数型一般是 1.2%、股票型一般是 1.5%，赎回费是 0.5%-1.5%，如果持有时间少于7天，赎回费就是1.5%，持有期多余7天，赎回费就是 0.5%（18年新出的政策），然后还要注意自己的证券公司手续费(华泰场内交易的买卖佣金是万分之二 0.02%)。所以一级市场申购然后到二级市场卖出的溢价套利总手续费就在 1.22%--1.51% 之间，只要溢价率超过 1.51%，那么理论上就能进行套利。
- **华宝合作：用华宝证券就只有 0.15%，同样场内申购1万块，华泰要交150元手续费，华宝只交15元**
- 开通后：溢价套利的总手续费就降低到了 0.13%--0.16% 之间

### 4.2. 风险
- 前提：基金场外净值是1元，场内交易价格是2元，所以以1元的净值进行申购，但是 T日 申购的要在 T+2 日才能卖出，假如在此期间场内交易价格暴跌，由2元跌到了0.9元，那不就成了我1元买，0.9元卖，这就导致套利失败了
- 折溢价率越大越好，折溢价率越大，安全空间就越大，利润空间也越大

- 红色的正数代表溢价，绿色的负数代表折价

> 1. 折溢价率要大，至少得是三四个点以上
> 2. 成交额要大，最好是几百万以上，方便出手

## 5. 如何套利
### 5.1 无底仓套利
#### 5.1.1 溢价套利
- 假设有支 LOF 基金，场内价格是 2 元，场外基金净值是 1 元，这个时候就是溢价，溢价率为（2-1）/ 1 = 100%
  - 套利：以 1 元的净值从场外买入此基金，然后以 2 元的价格从场内卖出
> 具体操作是：打开华泰证券，找到场内基金--场内申购，然后输入基金代码和申购金额，点击确定即可申购成功

![溢价套利操作1](溢价套利操作1.png)
![溢价套利操作2](溢价套利操作2.png)
![溢价套利操作3](溢价套利操作3.png)

- 开放式基金申购采取未知定价法，成交价格以当天收盘后的基金净值为准。
- 我们 T 日申购的基金会在 T+1 日到账（一般得在 T+1 日的晚上才能在持仓中看到），但当天不能交易，等到 T+2 日基金份额解冻后，即可在场内像卖股票一样直接卖出。整个溢价套利的流程就这样结束了~

#### 5.1.2. 折价套利
- 场内价格低于场外基金净值就叫做折价，比如力哥基金场内价格是0.5元，场外基金净值是1元，这个时候怎么操作呢？很简单，以0.5元的价格买入，然后以1元的净值卖出，低买高卖，成功获利~
> 具体操作是：打开华泰证券，在场内直接买入此基金，买入的基金份额会在 T+1 日到账，在份额到账后我们就可以进行场内赎回、以场外的净值卖出，场内赎回的资金会在 T+2 日到账

![折价套利操作1](折价套利操作1.png)
![折价套利操作2](折价套利操作2.png)
![折价套利操作3](折价套利操作3.png)

- 手续费
  - 场内买基金万一，之前LOF基金的赎回费是 0.5%，但 2018 年新出了政策，如果持有不到7天，那么就收 1.5% 的赎回费，我们是无底仓套利，而且时间很短，所以大概率会收 1.5% 的赎回费，也就是说折价套利的总手续费是 1.51%（根据场内费用计算，每个人会不同）。
  - 刚才说了溢价套利的总手续费才 0.13% -- 0.16%，折价套利是手续费是溢价套利的十几倍了，所以现在基本上只推荐溢价套利。

### 5.2. 底仓套利
- 下面讲的套利方法是在持有底仓的情况下进行的

#### 5.2.1. 溢价套利
##### 5.2.1.1 激进型**溢价**套利
> 具体操作是：卖出全部所持华宝油气份额（提醒：本来应该卖出全部持仓，但是！从19年8月9号开始，华宝油气单日最高申购 1000 元，也就是说我们最多只能拿 1000 元来套利，因此，我们现在只能卖出 1000 元，等限额解除后就可以不受此限制了），然后在收盘前通过场内申购相同市值的华宝油气。场内申购如下图，由于华宝油气是 QDII 基金，所以 T+2 日确认份额，T+3 日可交易。

##### 5.2.1.2. 保守型**溢价**套利
> 具体操作是：先确认自己的华宝油气市值，然后通过华泰证券申购同等市值的华宝油气（提醒：同样因为申购限额问题，最多只能申购1千块），**第二天**开盘时把持有的场内份额卖出。

##### 两种方式区别区别
- 激进型套利是**当天**先卖出所有华宝油气，**当天**重新申购同等市值的华宝油气
  - 本质是以 A 股的价格先卖出所有华宝油气，然后以今晚美股的收盘价买回所有华宝油气
  - 因为华宝油气跟踪的是美股油气指数，人家是北京时间的晚上开盘，我们当天的场内申购其实是按当晚美股收盘后的净值进行申购的
  - 如果当晚 XOP 暴跌，那我们申购的成本就更低，赚得就会更多；如果当晚 XOP 暴涨，那么我们就有亏损风险
- 保守型套利是**当天**先申购华宝油气，**第二天**才把原有份额全部卖出
  - 当晚 XOP 大涨，我原本持有的华宝油气肯定也会跟着大涨，虽然申购的那份是以较高的净值成交的，但是我卖出的价格更高，同样赚
    - **风险**：如果第二天场内华宝油气不涨反跌，导致溢价抹平，我卖出的价格不高而申购净值高就会套利失败。（概率比价低，大多情况和 xop 同涨同跌）
  - 当晚 XOP 大跌，意味着你申购的那份是以较低的净值成交的。因为我还持有一份场内的华宝，那么第二天开盘后场内华宝大概率也是下跌的，只要第二天华宝下跌的幅度小于等于 XOP 的下跌幅度，那么依然是有溢价空间的。
    - **风险**：第二天华宝的降幅远远大于XOP的降幅，比如XOP跌了4%，但是华宝跌了8%，把溢价抹平了，那套利就失败了（概率不大，涨跌幅应该比较接近的）

- 补充说明：华宝油气的涨跌情况和 XOP 很接近，因为这两者都是跟踪油气开采指数，只不过 XOP 是美股的 ETF，我们可以从尊嘉证券上查到 XOP 的波动情况。

---

#### 5.2.2. 折价套利
- 具体操作是：通过**场内赎回**，将所持所有份额赎回，同时在场内买入同等市值的基金

---

#### 趋势
- 一般在牛市中，大家都在场内买买买，容易推高场内的价格，导致**场内价格 > 场外净值**，这个时候**溢价套利**的机会就比较多
- 反之在熊市中，很多基民都在逃离股市，拼命卖卖卖，这就容易拉低场内的价格，导致**场内价格 < 场外净值**，这个时候**折价套利**的机会就比较多

## 6. 货基套利
- 在平时很多投资者都会把闲钱放在场内货基里，但是如果当市场资金紧张，或者出现投资机会，需要用现金的时候就需要你卖出持有的场内货基，所以你卖我卖大家卖就会拉低货基的场内价格，最终导致**场内价格 < 场外净值**

### 6.1. 步骤
- 先以场内的价格买入，然后再以场外的净值赎回，那我不就赚了
  - 1. 在证券账户上输入对应的代码买入货基
  - 2. 然后在当天收盘前赎回

![货币基金套](货币基金套利.png)

- 注意点
  - 在华泰交易 -- 更多，点击"货币基金"，选择"基金赎回"
  - 赎回代码是基金代码 +1，比如说 511810 的赎回代码是 511811
  - 一个是场内基金是 100 份起，场内货基的价格都在 100 左右，所以场内货基最低买入金额得在 1 万块左右
  - 在做货基套利的时候尽量选规模大、成交量活跃的基金
  - **风险**: 如果做套利的人非常多，因为你赎回、我也赎回，所以基金公司有可能暂停赎回或限额，基金规模越大，这种情况发生的概率就越低!

### 6.2. 计算收益
- 以华宝添益为例，假设现在的价格是 99.950，而且你在上面的截图上也能看到华宝添益的七日年化收益率 2.255%
- 假如今天以 99.950 买入，然后申请赎回，赎回的净值是以 100 计算（货币基金的净值是固定的，不会有波动）。赎回是 T+1 确认，T+2 到账。货币基金 T 日买入就计算利息，于是我还有额外的一天利息。

```
总收益 = 货币基金一天利息 + 货币基金价格与净值的差值
= 10000 × 2.255% ÷ 365×1 +（10000-9995）
= 0.618 + 5 = 5.618
也就是二天赚 5 块多，绝对收益率在 0.05% 左右

年化收益率 =（总收益 ÷ 2 × 365 ÷ 总花费资金）×100%
= 5.618 ÷ 2 × 365 ÷ 9995 × 100% = 10.26%
```

## 思维导图是什么？
- 思维导图又叫心智导图，也叫脑图
- 是表达发散性思维的有效图形思维工具，它是简单却又很有效的实用思维工具

## 思维导图有什么作用？
- 在学习上更容易记住知识点
- 在工作上帮助你做好工作规划、目标管理，在生活上让你过得有条理
- 比方说筹备婚礼、旅行安排、减肥/健身计划

## 工具
- 市面上思维导图的工具，比如 Xmind、Mindmanager、FreeMind、百度脑图，有免费的有付费的

## 使用工具画图（xmind 为例）
- 以香帅老师的 "为什么说基准利率是经济冷热的调节器" 文章为例，来把思维导图笔记写出来

1. 第一步：阅读知识
   - 消化知识：思维导图的质量很大程度上取决于我们的消化程度，所以一定要重视
   - 阅读过程中，把一些关键字、重点语句、知识点描述、例子等内容先记录或标记下来（思维导图的素材）
2. 第二步：绘制思维导图
   - 画思维导图，是一个梳理知识、关联知识点、加深理解知识的过程
   
### 绘图
- "中心主题"：为什么说基准利率是经济冷热的调节器

> 文章的开头，作者提到了"存款准备金"和"基准利率"是国家调控国家控制社会的信用规模和经济冷热的主要手段

- "存款准备金"并不是该文章讨论的重点，因此不展开内容，所以以"自由主题"的形式存在（双击空白处就会自动创建）
- "基准利率"是文章的重点概念，知识点和例子都是围绕该概念展开的，所以在中心主题加上一个主题

- 接下来，重点把"基准利率"的相关概念说清楚

- 文章接着讲述了"基准利率"的两个调控方向：加息和降息。我们就增加这两个子主题

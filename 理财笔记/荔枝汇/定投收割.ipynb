## 清楚定投目标
- 长期布局、长期定投可不是为了赚这五六个点

## 两种收割情况
### 仍处于布局阶段
- 还处于定投布局阶段的，更建议按照市值恒定法来进行操作，市值恒定法本身就能自动提醒我们进行收割。

![image](page1.png)

> 示例：
  - 假如小明未来一年打算在医药中投入50万，一年有50周，那么用50万÷50周=1万/周，也就是说理论上你每周要投入1万的资金，这样一年过后你正好把50万的资金全部投入了。
  - 把这个操作回顾一下，就是先确定你对某只基金的投资总额，然后用投资额除以计划定投的周期数就得出了每周理应定投的金额。
  - 假设说小明每周定投1万。那第一周目标市值就是1万，第二周就是2万，第三周3万，以此类推，第50周的时候就是我们计划投资总额50万。
  - 定投的时候，看我们账户里的实际市值和目标市值的差距是多少。比如说第一周。目标市值是1万，实际市值呢，也是1万，正好，那就不买不卖。第二周，我们目标市值是2万，实际市值的是1万8，少了两千，那我就买两千把它补成两万。第三周，我的目标市值是3万，但我的实际市值呢，是3万2，已经超过这个目标市值了，我就可以卖两千,落袋为安。

### 已经完成存量资金定投
- 主要用网格收割了

> 示例
  - 按照市值恒定法定投50周之后，你现在手中持有的医药市值刚好是50万，如果此时你不打算再追投新资金了，那你可以设置一个比例，当医药涨跌超过这个比例时你就可以进行买卖。
  - 比如说，你设置涨幅超过 5% 时就收割，跌幅超过 5% 时则买入。50万 × 5% = 2.5万，那么如果医药上涨，导致你的市值达到 52.5 万了，那么你就可以卖出 2.5万，把市值降低至 50万，这就是收割；
  - 重点说一点，就是收割、补仓的比例该设置为多少，一般来说设置为 2%-20% 之间其实都可以。这个比例的选择可以依据你是否有充足的时间来操作，因为你设置的比例越低那么达到你标准的频率就越多，那么你的买卖操作就越频繁。所以如果你时间充裕，那么可以设置稍微小一点的比例，比如 3%、5% 等等。如果你工作比较忙，不能经常看市场行情的可以设置稍微大一点，比如 15%、20% 等等。
  - 另外，我们一般建议单次买卖的金额要大于 1000 元，也就是说最好你的目标市值 × 设定的比例 ＞1000元，这是为了尽可能地降低手续费。
